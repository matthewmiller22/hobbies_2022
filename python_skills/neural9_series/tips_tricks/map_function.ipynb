### What's the Map function do? 

Simply put: It maps a **function** to an **iterable**. Syntax is: 

    (1) map(function, iterable)
    
To return the iterable as a list, simply wrap your *map* call with **list( )**
    
**Note: You may also (even, more preferably) use a list comprehension instead**



```python
numbers = [14,23,8,12,2,5,90]

def square(num):
    return num**2

list(map(square, numbers))
```
