### What is Threading? 

Threads are lightweight processes that allow us to run multiple tasks in parallel to speed up our scripts.

By defining various threads, we can run them simultaneously, in parallel. This is different than running processes sequentially 

### Try a trivial example to get the hang of syntax

```python
import threading

#initialize a function for testing
def helloworld():
    for x in range(10):
        print('Hello world!')

#initialize your various threads as .Thread() objects, and
#then start them running as appropriate

t1=threading.Thread(target=helloworld)
t1.start()
```

### Now show how you might use this for sequential functions

Instead of executing function1 and function2 *sequentially* below, they'll be run in parallel below:

```python
def function1():
    for x in range(1000):
        print("ONE")
        
def function2():
    for x in range(1000):
        print("TWO")
        
t1=threading.Thread(target=function1)
t2=threading.Thread(target=function2)

t1.start()
t2.start()
```

### What if we only want to initiate a process once a given thread has finished? 

In that case, simply use the **.join( )** method. E.g., below, if you only want "now print" to display *after* the thread has completed, do the following: 

```python
t1=threading.Thread(target=helloworld)
t1.start()
t1.join()

print('now print')
```

```python
t1=threading.Thread(target=helloworld)
t1.start()
#t1.join()

print('now print')
```
