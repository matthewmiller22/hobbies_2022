### Let's first review \*args and **kwargs

Reminder: \*args and \*\*kwargs allows your function to accept a variable number of parameters. 


\*args implies that things are beinb passed in list form, whereas \*\*kwargs implies a dictionary. 

Note that this is a form of argument parsing. 

```python
def myfunction(*args,**kwargs):
    print(args[0])
    print(args[1])
    print(args[2])
    
    print(kwargs['KEY0'])
    print(kwargs['KEY1'])
```

```python
myfunction('hi',3,True,KEY0='x',KEY1='y')
```

### Ok. Now actually onto the sys module. 

An important thing to note is that we can define variables from the command line, for instance as follows:

```python
import sys

filename=sys.argv[1]
message=sys.argv[2]
```

In the above, the first command line argument becomes stored as "filename" and the second command line argument becomes stored as "message"
