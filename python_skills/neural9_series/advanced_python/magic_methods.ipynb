```python
class Vector:
    def __init__(self,x,y): #make a 2D vector
        self.x=x
        self.y=y
        
    def __add__(self,other):
        return Vector(self.x+other.x,self.y+other.y)
    
    
    def __repr__(self): #how do you represent this class?
        return f"X {self.x}, Y {self.y}"
    

    
```

```python
v1 = Vector(10,20)
v2=Vector(50,60)
```
