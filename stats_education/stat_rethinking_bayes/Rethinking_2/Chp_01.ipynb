### Owls

The joke goes: Steps to Drawing an Owl: (1) Start off with some circles. (2) Draw the rest of the owl. 

The joke is that there are tons of steps from the basics to the final finished product (in our case, meaning, a great statistical model). In this course, we're going to learn all the little steps involved in drawing the owl! 

*Steps to Drawing the Bayesian Owl:*

    1. Theoretical Estimand: Let's clearly elaborate what the heck it is that we are trying to answer? We too often get lost even at thsi early step. 
    
    2. Scientific (causal) models: Let's discuss which process models are likely to be relevant for our data. 
    
    3. Use 1 + 2 to build statistical models. 
    
    4. Simulate from step 2 to validate that 3 yields 1. This way, we can validate that our software works! 
    
    5. Analyze real data

### Benefits of the Bayesian Approach

       1. The Bayesian approach is permissive and flexible. Can be applied to many different scientific models. 
   
       2. It can be used to express uncertainty at all levels. Including measurement error, sampling bias, etc. 
   
       3. It has direct solutiosns for measurement error and missing data (stay tuned for the latter part of the course).

## DAGs

How do we connect causal models to statistical procedures? Well: for statistical models to produce **scientific** insight, they require additional scientific (read: *causal*) models. 

The **reasons** for a statistical analysis are not found in the data, but rather their causes. 

**No causes in, no causes out**

### What is causal inference?

It is the attempt to understand the **causal (scientific)** model. Other ways to think of it:

    1. Causal inference is prediction of intervention. Remember: knowing a cause means being able to predict the consequences of an intervention OR knowing the counterfactual outcomes. 
    
        a. Does the wind blow the trees, or do the trees blow the wind?
    
    2. Causal inference is imputation of missing observations
    
Causal inference is both of these things (and more!)
