### III. How do I drill a pilot hole? 

Problem: Driving nails and screws can crack wood. 

Solution: Pre-drilling a **pilot hole** can help you to drive nails/screws without cracking wood : ) 

<ins> **Screws**: How do I choose the right-size drill bit for my pilot hole? </ins>

<ins>A: Choose a bit the **same size** as the **inner diameter**

<ins> **Nails**: How do I choose the right-size drill bit for my pilot hole? </ins>

A: Choose a bit **slightly smaller** than the **shank**

<ins>Note:</ins> You can also use **tape** on particularly fine piece of wood in order to avoid splintering. 

### IV. How do I cut glass?

I. Place glass on a towel in order to work safely

II. Mark your cut with a **wax pencil**

III. **Score** the glass with a **handheld glass cutter**. You should hear a slight "ripping sound", which will indicate a clean cut!

IV. Flip the glass piece over and snap off the part you'd like to cut!

### VI. What hammer should I use?

I. Curve claw hammer: Good all purpose choice!

II. Framing hammer: Longer and heavier for more power!

III. Ball peen: For flattening or curving metal

IV. Rubber mallet: Lighter touch for more delicate work

V. Sledge hammer: For wrecking stuff : ) 

**Note:** Other specialty hammers (e.g. drywall hammers, etc.) *do* exist for other jobs.

### VIII. How do I drill large holes? 

**Spade** bits can drill a large hole in one step

**Hole saws** come in even larager sizes for even bigger holes.

<ins>Spade Bit Drilling </ins>

I. Start by drilling a pilot hole with 1/8 inch bit

II. Drill from slow to fast speed, increasing as you go

III. Stop when the tip of the spade bit comes out the other side

IV. Then, flip the piece over and finish drilling *from the other side* for a clean hole 

<ins>Hole Saw Drilling </ins>

Repeat steps I-IV as above, but set your **center bit** to protrude 3/8 inch.
